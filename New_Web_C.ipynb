```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url='https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_2_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_2_0_na_na_na&as-pos=2&as-type=TRENDING&suggestionId=mobiles&requestId=3b44c040-c0dc-4f18-b039-76da72b4c0d8f'
```

```python
Header={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36 Edg/130.0.0.0','Accept-Language':'en-US'}
```

```python
page = requests.get(url,headers=Header)
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
titles=[]
def get_title(soup):
    for i in range(len(title)):
        tit=soup.find_all('div',class_='KzDlHZ')
        t=tit[i].text.split('(')[0].strip()
        titles.append(t)
```

```python
get_title(soup)
```

```python
titles
```

```python
img_patths=[]
def get_images(soup):
    for i in range(len(title)):
        img_url=image[i].find('img')['src']
        img_response = requests.get(img_url)
        img_filename = f"Images_scrapped/{titles[i].replace(' ', '_')}.jpg"
        img_patths.append(img_filename)
        with open(img_filename, 'wb') as img_file:
            img_file.write(img_response.content)
get_images(soup)
```

```python
img_patths
```

```python
prices_=[]
def get_prices(soup):
    prices=soup.find_all('div',class_='Nx9bqj _4b5DiR')
    for i in range(len(title)):
        prices_.append(prices[i].text)
    
```

```python
get_prices(soup)
```

```python
prices_
```

```python
doc={'Title':titles,'PRICE':prices_,'Img_path':img_patths}
```

```python
df=pd.DataFrame.from_dict(doc)
```

```python
df
```

```python
df.to_csv('flipkart.csv')
```
